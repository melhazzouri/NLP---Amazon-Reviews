# Import the Libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Retrieve HTML data and Extract the Data Elements

```python
search_query="phone+Sumsung"
```

```python
base_url="https://www.amazon.com/s?k="
```

```python
url=base_url+search_query
url
```

```python
header={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36','referer':'https://www.amazon.com/s?k=phone+Sumsung'}
```

```python
search_response=requests.get(url,headers=header)
```

```python
search_response.status_code
```

```python
search_response.text
```

```python
cookie={} # insert request cookies within{}
def getAmazonSearch(search_query):
    url="https://www.amazon.com/s?k="+search_query
    print(url)
    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchasin(asin):
    url="https://www.amazon.com/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):
    url="https://www.amazon.com"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

# Loop to get the content of the page of the required page

```python
product_names=[]
for i in range (1,7):
    response=getAmazonSearch('phone+Sumsung')
    soup=BeautifulSoup(response.content)
    for j in soup.findAll("span",{'class':'a-size-medium a-color-base a-text-normal'}): # the tag which is common for all the names of products
        product_names.append(j.text) #adding the productames to the list
```

```python
product_names
```

```python
len(product_names)
```

# Loop to get all the ASIN numbers of the each product in each page

```python
data_asin=[]
for i in range(1,7):
    response=getAmazonSearch('phone+Sumsung')
    soup=BeautifulSoup(response.content)
    for j in soup.findAll("div",{'class':"sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 AdHolder sg-col s-widget-spacing-small sg-col-12-of-16"}):
        data_asin.append(j['data-asin'])
```

```python
data_asin
```

```python
len(data_asin)
```

# Loop for the links of all the products ASIN numbers



```python
link=[]
for i in range(len(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
len(link)
```

# Loop to collect all the reviews of every data-asin number

```python
reviews=[]
for j in range(len(link)):
    for k in range(100):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            reviews.append(i.text)
```

```python
len(reviews)
```

# Create the Data Frame

```python
rev={'Reviews':reviews}
```

```python
reviews = pd.DataFrame.from_dict(rev)
pd.set_option('max_colwidth',70)
```

```python
reviews
```

# Convert the dataframe into CSV file 

```python
reviews.to_csv('SumsungReviews.csv')
```
